```python
import pandas as pd
import matplotlib.pyplot as plt

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")

df_cleaned = df.dropna(subset=["Age", "Salary"])

plt.plot(df_cleaned["EmployeeID"], df_cleaned["Salary"], label="Salary")
plt.plot(df_cleaned["EmployeeID"], df_cleaned["Age"], label="Age")

plt.legend()
plt.show()
```

```python

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df = df.dropna(subset=["Age", "Salary"])


plt.scatter(df["Age"],df["Salary"], color='red')

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()

```

```python
dept_counts = df["Department"].fillna("Unknown").value_counts()

plt.figure(figsize=(8,6))
plt.pie(dept_counts,
        labels=dept_counts.index,
        autopct='%1.1f%%')

plt.title("Department Distribution")
plt.show()
```

```python

dept_counts = df["Department"].value_counts()
plt.bar(dept_counts.index, dept_counts.values)

plt.xlabel("Department")
plt.ylabel("Count")
plt.title("Department Distribution")

plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.hist(df["Salary"].dropna(), bins=10, color='green', edgecolor='black')

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.bar(df["EmployeeID"],df["Salary"])
plt.title("EmployeeID")
plt.xticks(rotation=45)
plt.subplot(1,2,2)
plt.bar(df["EmployeeID"],df["Age"])
plt.title("Age")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
sns.barplot(
    x="Department",
    y="Salary",
    data=df
)
plt.show()
```

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(8,4))
sns.barplot(data=df, x="Department", y="Age")

plt.title("Average Age by Department")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[["Age","Salary"]].corr()
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=1
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and perform initial cleaning for 'Age' and 'Salary'
df = pd.read_csv("/content/drive/MyDrive/large_uncleaned_dataset.csv")
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
df = df.dropna(subset=["Age", "Salary"])

sns.pairplot(
    df[
        ["Age","Salary"]
    ]
)

plt.show()
```

```python
sns.pairplot(
    df,
    vars=["Age","Salary","EmployeeID"]
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))

sns.violinplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Salary Distribution by Department")
plt.show()
```

```python

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

plt.figure(figsize=(8,5))

sns.violinplot(
    data=df,
    x="Department",
    y="Age"
)

plt.title("Age Distribution by Department")
plt.show()
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="Department"
)

plt.title("Employee Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python

plt.figure(figsize=(7,5))

ax = sns.countplot(
    data=df,
    x="Department"
)

plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```
