```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/netflix_titles.csv")

print(df.head())
```

```python
print(df.shape)        # Rows and Columns
print(df.columns)      # Column Names
print(df.info())       # Data Types
print(df.describe())   # Numerical Summary
print(df.describe(include='object'))  # Categorical Summary
```

```python
print(df.isnull().sum())

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
print("Duplicates:", df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")

df["date_added"] = pd.to_datetime(df["date_added"], format='mixed')
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x="type")

plt.title("Movies vs TV Shows")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="rating",
    order=df["rating"].value_counts().index
)

plt.title("Content Ratings")
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.histplot(
    df["release_year"],
    bins=30,
    kde=True
)

plt.title("Release Year Distribution")
plt.show()
```

```python
top_country = df["country"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_country.values,
    y=top_country.index
)

plt.title("Top 10 Countries")
plt.show()
```

```python
genres = df["listed_in"].str.split(", ").explode()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

plt.title("Top 10 Genres")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x="rating",
    hue="type"
)

plt.xticks(rotation=45)
plt.title("Rating vs Content Type")
plt.show()
```

```python
df["year_added"] = df["date_added"].dt.year

plt.figure(figsize=(12,5))

sns.countplot(
    data=df,
    x="year_added",
    order=sorted(df["year_added"].dropna().unique())
)

plt.xticks(rotation=45)
plt.title("Content Added to Netflix by Year")
plt.show()
```

```python
Q1 = df["release_year"].quantile(0.25)
Q3 = df["release_year"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[
    (df["release_year"] < lower_bound) |
    (df["release_year"] > upper_bound)
]

print("Number of Outliers:", len(outliers_iqr))
print(outliers_iqr[["title","release_year"]].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="release_year"
)

plt.title("Box Plot of Release Year")
plt.show()
```

```python
from scipy.stats import zscore

df["z_score"] = zscore(df["release_year"])

outliers_z = df[
    (df["z_score"] > 3) |
    (df["z_score"] < -3)
]

print("Number of Outliers:", len(outliers_z))
print(outliers_z[["title","release_year","z_score"]].head())
```

```python
df_no_outliers = df[
    (df["release_year"] >= lower_bound) &
    (df["release_year"] <= upper_bound)
]

print("Original Shape:", df.shape)
print("New Shape:", df_no_outliers.shape)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(x=df["release_year"])
plt.title("Before Removing Outliers")

plt.subplot(1,2,2)
sns.boxplot(x=df_no_outliers["release_year"])
plt.title("After Removing Outliers")

plt.show()
```

```python
bins = [1900, 2000, 2010, 2020, 2030]

labels = [
    "Before 2000",
    "2001-2010",
    "2011-2020",
    "2021+"
]

df["year_group"] = pd.cut(
    df["release_year"],
    bins=bins,
    labels=labels
)

df[["release_year", "year_group"]].head()
```

```python
sns.countplot(
    data=df,
    x="year_group"
)

plt.title("Content by Release Year Group")
plt.show()
```

```python
# ==========================================
# NETFLIX DATASET - COMPLETE EDA WORKFLOW
# ==========================================

# STEP 1 : IMPORT LIBRARIES
# ==========================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import zscore
from sklearn.preprocessing import LabelEncoder

sns.set_style("whitegrid")

# STEP 2 : LOAD DATASET
# ==========================================

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/netflix_titles.csv")

print("Dataset Loaded Successfully")
print(df.head())

# STEP 3 : UNDERSTAND DATASET
# ==========================================

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nInfo:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

# STEP 4 : DATA CLEANING
# ==========================================

df.drop_duplicates(inplace=True)

df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")

df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce"
)

print("\nCleaning Completed")

# STEP 5 : OUTLIER DETECTION
# ==========================================

# Box Plot

plt.figure(figsize=(8,5))

sns.boxplot(
    x=df["release_year"]
)

plt.title("Box Plot - Release Year")
plt.show()

# IQR Method

Q1 = df["release_year"].quantile(0.25)
Q3 = df["release_year"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr = df[
    (df["release_year"] < lower) |
    (df["release_year"] > upper)
]

print("IQR Outliers:", len(outliers_iqr))

# Z Score Method

df["zscore"] = zscore(df["release_year"])

outliers_z = df[
    (df["zscore"] > 3) |
    (df["zscore"] < -3)
]

print("Z-Score Outliers:", len(outliers_z))

# STEP 6 : FEATURE ENGINEERING
# ==========================================

# Binning

bins = [1900,2000,2010,2020,2030]

labels = [
    "Before 2000",
    "2001-2010",
    "2011-2020",
    "2021+"
]

df["year_group"] = pd.cut(
    df["release_year"],
    bins=bins,
    labels=labels
)

# Label Encoding

le = LabelEncoder()

df["type_encoded"] = le.fit_transform(
    df["type"]
)

# One Hot Encoding

df_encoded = pd.get_dummies(
    df,
    columns=["type"],
    drop_first=True
)

print("Feature Engineering Completed")

# STEP 7 : UNIVARIATE ANALYSIS
# ==========================================

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="type"
)

plt.title("Movies vs TV Shows")
plt.show()

# STEP 8 : BIVARIATE ANALYSIS
# ==========================================

plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="rating",
    hue="type"
)

plt.xticks(rotation=45)

plt.title("Rating vs Content Type")

plt.show()

# STEP 9 : MULTIVARIATE ANALYSIS
# ==========================================

top_country = df["country"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_country.values,
    y=top_country.index
)

plt.title("Top Countries")

plt.show()

# STEP 10 : FINDINGS
# ==========================================

print("\nKEY FINDINGS")

print("""
1. Movies dominate Netflix content.

2. TV-MA is the most common rating.

3. United States contributes most content.

4. Netflix content increased rapidly after 2015.

5. International content is growing steadily.
""")

# STEP 11 : STORYTELLING
# ==========================================

print("\nDATA STORY")

print("""
Netflix has expanded significantly over
the last decade.

Movies remain the dominant content type,
while TV Shows continue to grow.

The platform is becoming increasingly global,
with strong contributions from India,
South Korea and the United Kingdom.

Most content targets mature audiences,
indicating strong demand in that segment.
""")

# STEP 12 : BUSINESS INSIGHTS
# ==========================================

print("\nBUSINESS INSIGHTS")

print("""
1. Increase investment in TV Shows.

2. Expand international content production.

3. Promote family-friendly content.

4. Focus on high-performing genres.

5. Continue global expansion strategy.
""")

# STEP 13 : CONCLUSION
# ==========================================

print("\nCONCLUSION")

print("""
The Netflix dataset reveals rapid content
growth, strong movie dominance, and increasing
international diversification.

The platform continues to expand its global
presence while serving a broad audience.
""")

print("\nPROJECT COMPLETED SUCCESSFULLY")
```
